SAVE THE DATA AS A JSON FILE

```python
import csv
import json
```

```python
file = open('framingham.csv')
csvreader = csv.reader(file)
```

```python
header = []
header = next(csvreader)
# header

my_list_of_dictionary_csv = []
for row in csvreader:
    my_list_of_dictionary_csv.append(row)
```

```python
my_list_of_dictionary = []
a = 1
for row in my_list_of_dictionary_csv:
    my_dict = {}
    my_dict['record_count'] = a
    a += 1
    for i in range(len(row)):
        my_dict[header[i]] = row[i]
    my_list_of_dictionary.append(my_dict)
```

```python
my_list_of_dictionary
```

```python
# Serializing json 
json_object = json.dumps(my_list_of_dictionary, indent = 4)
  
# Writing to sample.json
with open("json_response.json", "w") as outfile:
    outfile.write(json_object)
```
